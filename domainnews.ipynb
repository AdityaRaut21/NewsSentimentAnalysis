```python
import requests
import json
from datetime import datetime, timedelta
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
from transformers import pipeline

# Download necessary NLTK data
nltk.download('vader_lexicon', quiet=True)

# Load multilingual sentiment analysis model
multilingual_sentiment_analyzer = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

def get_news(api_key, category, query, language):
    one_week_ago = (datetime.now() - timedelta(days=7)).strftime('%Y-%m-%d')
    base_url = "https://newsapi.org/v2/everything"
    
    params = {
        'apiKey': api_key,
        'q': f"{category} AND {query}",
        'language': language,
        'from': one_week_ago,
        'sortBy': 'publishedAt'
    }
    
    response = requests.get(base_url, params=params)
    
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code}")
        return None

def translate_sentiment_to_hindi(sentiment):
    sentiment_map = {
        "Positive": "सकारात्मक 😊",
        "Negative": "नकारात्मक ☹️",
        "Neutral": "तटस्थ 😐"
    }
    return sentiment_map.get(sentiment, sentiment)

def analyze_sentiment(text, language):
    if language == 'en':
        sia = SentimentIntensityAnalyzer()
        sentiment_scores = sia.polarity_scores(text)
        compound_score = sentiment_scores['compound']
        
        if compound_score >= 0.05:
            return "Positive", sentiment_scores
        elif compound_score <= -0.05:
            return "Negative", sentiment_scores
        else:
            return "Neutral", sentiment_scores
    elif language == 'hi':
        result = multilingual_sentiment_analyzer(text)
        label = result[0]['label']
        score = result[0]['score']
        
        if "positive" in label.lower():
            return translate_sentiment_to_hindi("Positive"), {"confidence": score}
        elif "negative" in label.lower():
            return translate_sentiment_to_hindi("Negative"), {"confidence": score}
        else:
            return translate_sentiment_to_hindi("Neutral"), {"confidence": score}
    else:
        return "Unknown", {}

def summarize_text(text):
    # Simple rule-based summarization: if the text is long, only keep the first sentence.
    if len(text) > 150:
        return text.split('.')[0] + '.'
    return text

def main():
    api_key = "033302b4ad3c4ca1bc664e1c784bb622"  # Your API key
    categories = ["finance", "education", "healthcare"]
    languages = ['en', 'hi']
    
    for language in languages:
        for category in categories:
            print(f"\nFetching {language} news for {category.capitalize()}:")
            news = get_news(api_key, category, category, language)
            
            if news and news['articles'][:10]:
                for i, article in enumerate(news['articles'][:10], 1):
                    title = article['title']
                    description = article['description'] or "No description available"
                    description = summarize_text(description)
                    
                    title_sentiment, title_confidence = analyze_sentiment(title, language)
                    desc_sentiment, desc_confidence = analyze_sentiment(description, language)
                    
                    print(f"\nArticle {i}:")
                    print(f"Title: {title}")
                    print(f"Title Sentiment: {title_sentiment} (Confidence: {title_confidence.get('compound', title_confidence.get('confidence', 'N/A'))})")
                    print(f"Description: {description}")
                    print(f"Description Sentiment: {desc_sentiment} (Confidence: {desc_confidence.get('compound', desc_confidence.get('confidence', 'N/A'))})")
                    print(f"URL: {article['url']}")
                    print(f"Published at: {article['publishedAt']}")
                    print("---")
            else:
                print(f"No articles found for {category} in {language}")

if __name__ == "__main__":
    main()
```
